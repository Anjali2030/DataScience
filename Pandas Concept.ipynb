# 5. Plotting using Matplotllib

```python
# Data Visualization By Matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Use internal data set in this library
```

```python
# 2-D line plot graph
price=[50000,40000,10000,90000,85000,70000,75000]
year=[2010,2011,2012,2013,2014,2015,2016]
plt.plot(year,price)
plt.title('Graph of a mobile company compare the price ')
plt.xlabel('Year')
plt.ylabel('Price of Mobile ')
```

# Bar Chart

```python
Students=[10,20,30,15,30,45]
color=['Pink','Red','Blue','Orange','Purple','Yellow']
plt.bar(color,Students)
plt.title('Students favourite color')
plt.xlabel('Color')
plt.ylabel('Number of Students')
```

```python
Students=[10,20,30,15,30,45]
color=['Pink','Red','Blue','Orange','Purple','Yellow']
plt.barh(color,Students)
plt.title('Students favourite color')
plt.xlabel('Color')
plt.ylabel('Number of Students')
```
