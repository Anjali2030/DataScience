```python
# preprocessing
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from skimage.io import imread        # read an image also we can use open cv
from skimage.transform import resize # resize the image
target=[]
images=[]
flat_data=[]

DATADIR='/content/images'
CATEGORIES=['pretty sunflower','rugby ball','icecream cone']

for category in CATEGORIES:
    class_num=CATEGORIES.index(category) # label encoding the value
    path=os.path.join(DATADIR,category) # create a path to use all the images
    #print(path)
    for img in os.listdir(path):
        img_array= imread(os.path.join(path,img))
        #print(img_array.shape)
        #plt.imshow(img_array)
        img_resized=resize(img_array,(150,150,3)) # resize using skimage normalize the value 0 to 1
        flat_data.append(img_resized.flatten())
        images.append(img_resized)
        target.append(class_num)
flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(flat_data,target,test_size=0.2,random_state=2)
```

```python
# algorithm for classification data use Support vector algorithm
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
    {'C':[1,10,100,1000],'kernel':['linear']},
    {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
    ]
svc=svm.SVC(probability=True)
clf=GridSearchCV(svc,param_grid)
clf.fit(X_train,Y_train)
```

```python
Y_pred=clf.predict(X_test)
Y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(Y_pred,Y_test)
```

```python
confusion_matrix(Y_pred,Y_test)
```

```python
# Save the model using Pickle Library
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```
